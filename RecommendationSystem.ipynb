```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder
```

```python
products_data = pd.read_csv("/home/dennis/Desktop/regulusProducts.csv")
display(products_data.head())   
```

```python
products_data.describe()
```

```python
#binary encode the order status to get those bought and those not bought
df= pd.DataFrame(products_data)
display(df["Order status"].unique())
label_encoder = LabelEncoder()
df["Order status"] = label_encoder.fit_transform(df['Order status'])
display(df["Order status"].unique())
display(df)
```

```python
#create a pivot table
products_bought = df.pivot_table(values="Order status",index="Customer name", columns="Product")
display(products_bought)
```

```python
dictionary = products_bought["Drawing Book A3 Ref053"]
display(dictionary)
```

```python
similarBooks = products_bought.corrwith(dictionary)
similarBooks = similarBooks.dropna()
similarBooksdf = pd.DataFrame(similarBooks)
display(similarBooksdf)
```

```python
similarBooks.sort_values(ascending=False)
```

```python
product_stats = df.groupby('Product').agg({'Order status':[np.size, np.mean]})
display(product_stats)
```

```python
popular_products = product_stats["Order status"]["size"]>=1000
product_stats[popular_products].sort_values(by=("Order status","mean"),ascending=False)[:15]
display(product_stats)
```

```python
df = product_stats[popular_products].join(pd.DataFrame(similarBooks.reset_index(0),columns=["similarity"]), on="Product")
display(df.head())
```
